```python
%matplotlib inline
from skimage import novice, data, io, exposure
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


brainpic = novice.open('braincomplete.png')
brainpicarray = mpimg.imread('braincomplete.png','True') 


fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
img_adapteq = exposure.equalize_adapthist(brainpicarray, clip_limit=0.5)
p2, p98 = np.percentile(brainpicarray, (5, 70))
img_eq2 = exposure.rescale_intensity(brainpicarray, in_range=(p2, p98))



plt.subplot(221)
plt.imshow(img_adapteq)
plt.subplot(222)
plt.imshow(brainpicarray)
plt.subplot(223)
ht=plt.hist(img_eq2.ravel(), bins=256, histtype='step', color='black')
plt.subplot(224)
ht=plt.hist(brainpicarray.ravel(), bins=256, histtype='step', color='black')
```

#### We can see that  after changing the value of the clip limit, we can increase the contrast and the detailing of the output image, but simultaneously considering the fact that it does not white out the areas of the image which already had a good detail in it.

On performing **equlize_hist** on the Image, we observe that almost all the pixels are getting utilized to generate an enhanced image and there is no scope of further enhancement of user specific information. This method is a global method wherin performing any changes to view an area, it shall affect the entire image as well as its corresponding histogram. <br>Whereas in **equalize_adapthist**, we find that we have been provided with an option to vary the contrast of some specific portion which helps us to adjust the contrast as per our requirements. We have that widget to toggle and change the contrast of the image as per our needs and requirements which is a very handy and efficient tool. However, it does not affect the entire image and only area specific information is highlighted.
Thus, equalize_adapthist is far better than equalize_hist 

When we increase the Clip Limit to **0.5**, we observe that higher values give more contrast to the image.
